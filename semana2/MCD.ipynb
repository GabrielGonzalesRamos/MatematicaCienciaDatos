````python
data = pd.read_csv("banca.csv", sep=";")
data.select_dtypes(["float64", "int64"])
````

Convirtiendo mi **DataFrame** n√∫merico en un **array**

````python
arrayBanca = data.select_dtypes(["float64", "int64"]).to_numpy()
arrayBanca
````

Viendo las dimensiones de mi **array**

````python
arrayBanca.shape
````

Calcule el vector de **medias**

````python
Medias = np.mean(arrayBanca, axis=0)
Medias
````

Calcule **la resta de la matriz** con el vector de **medias**

````python
Resta = arrayBanca - Medias
Resta
````

Calcule la matriz de **covarianza**

````python
Covarianza = np.cov(arrayBanca, rowvar=False)
Covarianza
````

Verificando la cantidad de columnas y filas que tiene mi variable ``Covarianza``

````python
Covarianza.shape
````

Nos piden calcular la matriz inversa del **array** ``arrayBanca`` 
 - Primero se validar√° si tiene **determinante**, para realizar lo solicitado

````python
Determinante = np.linalg.det(Covarianza)
Determinante
````

Al ver que tiene determinante, se procedera a calcular la **matriz inversa**

````python
Inversa = np.linalg.inv(Covarianza)
Inversa
````

````python
MatrizIdentidad = np.dot(Covarianza, Inversa)
MatrizIdentidad
````
